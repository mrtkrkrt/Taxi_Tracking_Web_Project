```python
import pandas as pd
import numpy  as np
```

```python
data = pd.read_csv("../input/taximovementconcatenated/allCars.csv", names= ["date", "lat", "lng", "id"])
data.head(5)
```

```python
vehicle_1 = data[data["id"] == 31].drop("id", axis = 1)
vehicle_2 = data[data["id"] == 41].drop("id", axis = 1)
vehicle_3 = data[data["id"] == 3].drop("id", axis = 1)
vehicle_4 = data[data["id"] == 122].drop("id", axis = 1)
```

```python
# Vehicle_1 data edit
vehicle_1_edited = vehicle_1[vehicle_1.date.str.startswith('2018-11-30')]
vehicle_2_edited = vehicle_2[vehicle_2.date.str.startswith('2018-11-30')]
vehicle_3_edited = vehicle_3[vehicle_3.date.str.startswith('2018-11-30')]
vehicle_4_edited = vehicle_4[vehicle_4.date.str.startswith('2018-11-30')]

print(np.array(vehicle_1_edited).shape)
print(np.array(vehicle_2_edited).shape)
print(np.array(vehicle_3_edited).shape)
print(np.array(vehicle_4_edited).shape)
```

```python
vehicle_1_edited.to_csv("vehicle_1.csv", sep=',', encoding='utf-8')
vehicle_2_edited.to_csv("vehicle_2.csv", sep=',', encoding='utf-8')
vehicle_3_edited.to_csv("vehicle_3.csv", sep=',', encoding='utf-8')
vehicle_4_edited.to_csv("vehicle_4.csv", sep=',', encoding='utf-8')
```
